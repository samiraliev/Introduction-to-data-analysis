```python
import numpy as np
import pandas as pd
```

```python
#Выкачиваем данные по аэропортам
columns = ['Airport ID', 'Name', 'City', 'Country', 'IATA', 
'ICAO', 'Latitude', 'Longitude', 'Altitude', 'Timezone', 
'DST', 'Tz database', 'Type', 'Source']
airports = pd.read_csv('C:/Users/1311959/Documents/R/airports.csv', 
                       names = columns)
airports.head()
```

```python
# отбор европейских стран, включая заморские территории
euro_air = airports[airports['Tz database'].str.contains('Europe', na=False)]
euro_air
```

```python
# Отбор европейских стран с нужными территориями
euro_airports = euro_air[(euro_air['Country'] != 'Russia') & 
                (euro_air['Country'] != 'Turkey') & 
                (euro_air['Country'] != 'United States') &
                (euro_air['Country'] != 'Guernsey') &
                (euro_air['Country'] != 'Jersey') &
                (euro_air['Country'] != 'Isle of Man') &
                (euro_air['Country'] != 'Gibraltar') &
                (euro_air['City'] != 'Beja (madeira)') &
                (euro_air['City'] != 'Funchal') &
                (euro_air['City'] != 'Melilla')]

euro_airports
```
